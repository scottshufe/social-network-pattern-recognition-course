# 导入相关Python包

```python
import networkx as nx
import matplotlib.pyplot as plt
% matplotlib inline
```

# 创建网络

```python
# Karate Club network

# 读入网络数据
edge_list = []
with open('karate_club_graph.txt', 'r') as f:
    lines = f.readlines()
    for line in lines:
        edge = tuple(line.replace('\n', '').split(' '))
        edge_list.append(edge)
```

```python
# 创建网络对象
# 加载边数据

G = nx.Graph()
G.add_edges_from(edge_list)
```

```python
# 查看网络的基本信息

print(nx.info(G))
```

```python
# 简单可视化

layout = nx.spring_layout(G)
nx.draw(G, pos=layout, with_labels=True)
```

# 网络节点的度分布

```python
# 获取所有节点的度
nx.degree(G)
```

```python
# 获取单个节点的度
G.degree('0')
```

```python
# 计算并绘制节点的度分布

# 计算节点的度分布
from collections import Counter

degree_list = []
for node in G.nodes():
    degree_list.append(G.degree(node))
```

```python
Counter(degree_list)
```

```python
degrees = list(sorted(Counter(degree_list).keys()))
values = [Counter(degree_list)[degree] for degree in degrees]
```

```python
degrees
```

```python
values
```

```python
plt.plot(degrees, values)
```

# 节点影响力度量——中心性计算

- Networkx包中心性计算参考链接：https://networkx.org/documentation/stable/reference/algorithms/centrality.html

1. <b>度中心性（Degree centrality）：</b>
    - 一个顶点的度中心性指的是与该顶点关联的其他顶点个数。通常来说，度中心性越大的顶点其重要性越大。在有向图中，度中心性可分为出度中心性，与入度中心性。
<br />
<br />
2. <b>介数中心性（Betweenness centrality）</b>
    - 介数中心性衡量了一个顶点出现在其他任意两个顶点对之间的最短路径的次数，也就是说，如果一个顶点出现在任意两个顶点间最短路径的次数越多，那么该顶点的介数中心性就越大，主要用于衡量一个顶点在图或网络中承担“桥梁”角色的程度。
<br />
<br />
3. <b>接近中心性（Closeness centrality）</b>
    - 接近中心性即一个顶点到其他所有顶点的最短距离之和的倒数。如果一个顶点到其他顶点的最短距离都很小，那么它的接近中心性就较高。相比介数中心性，接近中心性更接近几何上的中心位置。
<br />
<br />
4. <b>特征向量中心性（Eigenvector centrality）</b>
    - 特征向量中心性的基本思想是，一个节点的中心性是相邻节点中心性的函数。也就是说，与你连接的人越重要，你也就越重要。度中心性高的顶点特征向量中心性不一定高，因为所有的连接者有可能特征向量中心性很低；同理，特征向量中心性高并不意味着它的度中心性高。
<br />
<br />
5. <b>PageRank</b>
    - PageRank最初是Google公司所使用的对其搜索引擎搜索结果中的网页进行排名的一种算法，后来也成为了衡量有向网络中节点重要性的指标。其基本假设是：重要的页面往往更多地被其他页面引用，来自重要页面的跳转应该重于来自其他不受欢迎页面的跳转。
<br />
<br />

## 度中心性计算

 - nx.degree_centrality(G)：返回节点为键，度中心性为值的字典

```python
nx.degree_centrality(G)
```

```python
# 按度中心性从大到小排序输出

for i in sorted(nx.degree_centrality(G).items(), key=lambda x: x[1], reverse=True):
    print(i)
```

## 介数中心性计算

```python
nx.betweenness_centrality(G)
```

## 特征向量中心性计算

## PageRank

- nx.pagerank(G)：返回节点为键，PageRank得分为值的字典

```python
G_test = nx.binomial_graph(10, 0.3, directed=True)
layout = nx.spring_layout(G_test)
plt.figure(1)
nx.draw(G_test, pos=layout, node_color='y')

pr=nx.pagerank(G_test)
# print(pr)
for node, pageRankValue in pr.items():
    print("Node: %d, PageRank Value: %.4f" %(node, pageRankValue))

plt.figure(2)
nx.draw(G_test, pos=layout, node_size=[x * 6000 for x in pr.values()], node_color='m', with_labels=True)
plt.show()
```
